# KS Test for Strict Stationarity 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp
```

Synthetic Data Generation

```python
n= 500;

strict_stationarity_series = np.random.normal(0,1,n)
non_strict_stationarity_series = np.concatenate([np.random.normal(0,1,n//2),np.random.normal(0,2,n//2)])
```

# KS Test (Kolmogorov-Smirnov Test)

```python
from scipy.stats import ks_2samp
def ks_test_stationarity(series):
    split = int(len(series)/2)
    series_first_half = series[:split]
    series_second_half = series[split:]
    stat, p_value = ks_2samp(series_first_half,series_second_half)
    print('KS Test Statistics Results')
    print(f'p_value: {p_value}')
    print(f'KS Statistic: {stat}')
    return stat, p_value 
#H0: Data is strict stationary
# H1: Data is not stationary 
```

```python
print("Strict Stationarity")
plt.figure(figsize=(12,12))
plt.subplot(211)
plt.plot(strict_stationarity_series)
plt.title(f"Strict Stationarity Series p-value: {ks_pvalue_strict}")
ks_stat_strict, ks_pvalue_strict = ks_test_stationarity(strict_stationarity_series)
# p-value greater than 0.05 implying that fail to reject null Hypothesis and that data is strict stationary

plt.subplot(212)
plt.plot(non_strict_stationarity_series)
ks_stat_non_strict, ks_pvalue_non_strict = ks_test_stationarity(non_strict_stationarity_series)
plt.title(f"Non-strict Stationarity p-value: {ks_pvalue_non_strict}")
# p-value less than 0.05 implying that fail to accept null Hypothesis and that data is strict stationary. This is also visible in the data series itself
```
